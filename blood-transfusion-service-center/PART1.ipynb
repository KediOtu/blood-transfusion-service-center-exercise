```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import fetch_openml
X,y = fetch_openml('blood-transfusion-service-center', version=1,return_X_y=True)

# In dataset, there are 5 features. V1, V2, V3, V4 and class. Target is class.

X_train,X_test,y_train,y_test=train_test_split(X,y) 
#The target attribute is a binary variable representing whether he/she donated blood in March 2007 (2 stands for donating blood; 1 stands for not donating blood).

plt.scatter(X_train['V1'],X_train['V2'])
plt.xlabel('V1')  #V1: Recency - months since last donation
plt.ylabel('V2')  #V2: Frequency - total number of donation
```

```python
plt.scatter(X_train['V1'],X_train['V3'])
plt.xlabel('V1')  #V1: Recency - months since last donation
plt.ylabel('V3')  #V3: Monetary - total blood donated in c.c
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

```python
predictions = model.predict(X_test)
print(predictions)
```
